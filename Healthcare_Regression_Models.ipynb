```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Simulating the data
np.random.seed(42)
n = 100

# Predictors
X1 = np.random.uniform(10, 100, n)  # Medication amount
X2 = np.random.uniform(1, 10, n)    # Follow-up frequency
X3 = np.random.uniform(0.5, 1, n)   # Treatment adherence

# Response variable (health outcome)
epsilon = np.random.normal(0, 5, n)
Y = 2 * X1 + 3 * X2 + 5 * X3 + epsilon  # True relationship

# Dataframe for visualization
data = pd.DataFrame({
    'Health Outcome': Y,
    'Medication Amount': X1,
    'Follow-Up Frequency': X2,
    'Treatment Adherence': X3
})

# Pairwise plots
sns.set(style="whitegrid")
sns.pairplot(data, kind='reg', plot_kws={'line_kws':{'color':'red'}})
plt.suptitle("Pairwise Relationships with Linear Regression Lines", y=1.02)
plt.show()

# Joint regression analysis
X = np.column_stack((X1, X2, X3))
model = LinearRegression().fit(X, Y)
coefficients = model.coef_
intercept = model.intercept_

print("Model Coefficients:")
print(f"Medication Amount (X1): {coefficients[0]:.2f}")
print(f"Follow-Up Frequency (X2): {coefficients[1]:.2f}")
print(f"Treatment Adherence (X3): {coefficients[2]:.2f}")
print(f"Intercept: {intercept:.2f}")

# Visualizing the contributions of each predictor
predictors = ['Medication Amount', 'Follow-Up Frequency', 'Treatment Adherence']
plt.bar(predictors, coefficients, color=['#440154FF', '#3B528BFF', '#21908CFF'])
plt.title('Impact of Predictors on Health Outcome')
plt.ylabel('Regression Coefficient')
plt.show()
```
